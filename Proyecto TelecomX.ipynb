```python
df_normalizado = pd.json_normalize(df_telecom.to_dict(orient='records'))
```

#🔧 Transformación

```python
pd.set_option('display.max_columns', None)
df_normalizado.head(35)
```

```python
df_normalizado['Churn'] = df_normalizado['Churn'].str.strip()
```

```python
df_normalizado['Churn'] = df_normalizado['Churn'].fillna('No')  # O cualquier valor que tenga sentido
```

```python
print(df_normalizado['Churn'].unique())
```

##*Chequeo de valores nulos*

```python
df_normalizado.isnull().sum()
```

##*Chequeo de valores duplicados*

```python
valores_duplicados = df_normalizado.duplicated().sum()
print(f"En el dataframe tenemos inicialmente un total de {valores_duplicados} valores duplicados")
```

```python
df_normalizado.drop_duplicates(inplace=True)
```

##*Chequeo si columnas numericas tienen strings*

```python
df_normalizado["customerID"] = df_normalizado["customerID"].astype(str)
df_normalizado["customer.gender"] = df_normalizado["customer.gender"].astype(str)
df_normalizado["customer.Partner"] = df_normalizado["customer.Partner"].astype(str)
df_normalizado["customer.Dependents"] = df_normalizado["customer.Dependents"].astype(str)
df_normalizado["account.Charges.Total"] = pd.to_numeric(df_normalizado["account.Charges.Total"], errors='coerce')
df_normalizado['account.Charges.Monthly'] = pd.to_numeric(df_normalizado['account.Charges.Monthly'], errors='coerce')
df_normalizado["customerID"] = df_normalizado["customerID"].astype(str)
df_normalizado["customerID"] = df_normalizado["customerID"].astype(str)
print(df_normalizado.dtypes)
```

```python
df_normalizado.head(10)
```

##*Creacion de la columna  'Cuentas_Diarias'*

```python
df_normalizado['Cuentas_Diarias'] = df_normalizado['account.Charges.Monthly'] / 30
df_normalizado['Cuentas_Diarias'] = df_normalizado['Cuentas_Diarias'].round(2)
df_normalizado[['account.Charges.Monthly', 'Cuentas_Diarias']].head()
```

#📊 Carga y análisis

##*Valores generales del Df a grandes rasgos*

```python
df_normalizado.describe().round(2)
```

##*Analisis Detallado*

```python
variables = ['customer.tenure', 'account.Charges.Monthly', 'account.Charges.Total', 'Cuentas_Diarias']

for var in variables:
    media = df_normalizado[var].mean()
    mediana = df_normalizado[var].median()
    desviacion = df_normalizado[var].std()
    minimo = df_normalizado[var].min()
    maximo = df_normalizado[var].max()

    print(f'--- {var} ---')
    print(f'Media: {media:.2f}')
    print(f'Mediana: {mediana:.2f}')
    print(f'Desviación Estándar: {desviacion:.2f}')
    print(f'Mínimo: {minimo:.2f}')
    print(f'Máximo: {maximo:.2f}\n')
```

##📈Grafico de Distribucion de Cuentas diarias en base al:
- monto diario y cantidad de clientes
- Cantidad de clientes a lo largo de los meses
- Monto Mensual segun la cantidad de clientes

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribución de Tenure (antigüedad)
sns.histplot(df_normalizado['customer.tenure'], bins=30, kde=True)
plt.title('Distribución de la Antigüedad (Tenure)')
plt.xlabel('Meses')
plt.ylabel('Cantidad de Clientes')
plt.show()


# Distribución de Cuentas Diarias
sns.histplot(df_normalizado['Cuentas_Diarias'], bins=30, kde=True, color='orange')
plt.title('Distribución de Cuentas Diarias')
plt.xlabel('Monto Diario')
plt.ylabel('Cantidad de Clientes')
plt.show()

# Distribución de Cargos Mensuales
sns.histplot(df_normalizado['account.Charges.Monthly'], bins=30, kde=True, color='green')
plt.title('Distribución de Cargos Mensuales')
plt.xlabel('Monto Mensual')
plt.ylabel('Cantidad de Clientes')
plt.show()
```

##*Informacion sobre la dispersion de los datos y forma del grafico*

```python
from scipy.stats import skew, kurtosis

for var in variables:
    asimetria = skew(df_normalizado[var])
    curtosis = kurtosis(df_normalizado[var])

    print(f'--- {var} ---')
    print(f'Asimetría: {asimetria:.2f}')
    print(f'Curtosis: {curtosis:.2f}\n')
```

##*⚖️Distribución de evasión de Clientes*

```python
conteo_churn = df_normalizado['Churn'].value_counts()
print(conteo_churn)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gráfico de barras para la distribución de churn
sns.countplot(data=df_normalizado, x='Churn', palette='Set2')
plt.title('Distribución de Clientes por Evasión')
plt.xlabel('Estado de Churn')
plt.ylabel('Cantidad de Clientes')
plt.show()
```

```python
# Gráfico de torta para ver proporciones
plt.figure(figsize=(6, 6))
plt.pie(conteo_churn, labels=conteo_churn.index, autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], startangle=90)
plt.title('Proporción de Clientes que Evasaron vs. Retenidos')
plt.show()
```

```python
# Cálculo de porcentaje
total_clientes = len(df_normalizado)
porcentaje_churn = (conteo_churn['Yes'] / total_clientes) * 100
porcentaje_no_churn = (conteo_churn['No'] / total_clientes) * 100

print(f'Clientes que se dieron de baja: {porcentaje_churn:.2f}%')
print(f'Clientes que permanecieron: {porcentaje_no_churn:.2f}%')
```

### 📌 Introducción
El presente análisis tiene como objetivo **explorar el fenómeno de evasión de clientes (Churn)** en una empresa de telecomunicaciones. La pérdida de clientes es uno de los principales desafíos en el sector, ya que impacta directamente en los ingresos y en la sostenibilidad del negocio a largo plazo.

A través de este estudio, buscamos:
- Identificar patrones de comportamiento asociados a la evasión.
- Detectar perfiles de clientes con mayor riesgo de cancelar el servicio.
- Proveer insights accionables que permitan desarrollar estrategias para reducir la tasa de churn.

---

El dataset inicial presentaba:
- Estructura anidada en formato JSON.

### ✅ Conclusiones e Insights
- Los **clientes con contratos mensuales son el grupo con mayor probabilidad de cancelar.**
- La **antigüedad es un factor crítico:** los clientes que llevan menos tiempo son más propensos a abandonar.

### 💡 Recomendaciones
**1. Incentivar contratos a largo plazo:**  
Ofrecer beneficios y descuentos atractivos para que los clientes elijan contratos anuales o bianuales.

**2. Programas de retención para nuevos clientes:**  
Diseñar estrategias de fidelización específicas para los primeros meses del contrato, cuando el riesgo de cancelación es más alto.

**3. Ofrecer incentivos para automatizar los pagos:**  
Promover el uso de métodos de pago automáticos mediante descuentos o beneficios exclusivos.

**4. Revisión de precios y calidad del servicio de fibra óptica:**  
Evaluar la competitividad del servicio de fibra óptica frente a la competencia para reducir la insatisfacción y las cancelaciones.

**5. Seguimiento proactivo:**  
Desarrollar alertas tempranas basadas en las características del cliente (corto tiempo de permanencia, cargos elevados, uso de fibra óptica) para intervenir antes de que el cliente decida cancelar.
